(1)应用全局直方图均衡化进行图像增强

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("TZB53.jpg", 0)
print(img.shape)

#直方图均衡化
imgH = cv2.equalizeHist(img)

#可视化
plt.subplot(121)
plt.imshow(img, "gray", vmin=0, vmax=255)
plt.subplot(122)
plt.imshow(imgH, "gray", vmin=0, vmax=255)
plt.show()
```

(2) 应用自适应直方图均衡化方法（AHE）进行图像增强

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("TZB53.jpg", 0)

def AHE(img, patchsize):
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(patchsize,patchsize))
    img_ahe = clahe.apply(img)
    return img_ahe

patchsize = 3
img_ahe1 = AHE(img, patchsize)

patchsize = 26
img_ahe2 = AHE(img, patchsize)

patchsize = 50
img_ahe3 = AHE(img, patchsize)

patchsize = 100
img_ahe4 = AHE(img, patchsize)

patchsize = 200
img_ahe5 = AHE(img, patchsize)

#figsize设置显示面板的宽和高
fig, ax = plt.subplots(2, 3, figsize=(10,7))
ax[0,0].imshow(img, "gray", vmin=0, vmax=255)
ax[0,0].set_title('Origin Image')
ax[0,0].axis('off')

ax[0,1].imshow(img_ahe1, "gray", vmin=0, vmax=255)
ax[0,1].set_title('patchsize=3')
ax[0,1].axis('off')

ax[0,2].imshow(img_ahe2, "gray", vmin=0, vmax=255)
ax[0,2].set_title('patchsize=26')
ax[0,2].axis('off')

ax[1,0].imshow(img_ahe3, "gray", vmin=0, vmax=255)
ax[1,0].set_title('patchsize=50')
ax[1,0].axis('off')

ax[1,1].imshow(img_ahe4, "gray", vmin=0, vmax=255)
ax[1,1].set_title('patchsize=100')
ax[1,1].axis('off')

ax[1,2].imshow(img_ahe5, "gray", vmin=0, vmax=255)
ax[1,2].set_title('patchsize=200')
ax[1,2].axis('off')

plt.show()
```

随着子块尺寸变小，局部区域变得更平滑;随着子块尺寸变大，局部区域就变得更粗糙.
